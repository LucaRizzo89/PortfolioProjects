### Parsing and formatting 'Salary Estimate' 

This column is probably the most important in our dataset, considering the purpose of the analysis. Therefore we will get rid of the rows with null values, since we don't have a way to extract these from any other part of the dataset.

```python
df['Salary Estimate'].value_counts().head(50)
```

Having a quick look at the unique values, we can see that there are no nulls, however, the column has 214 values of '-1' that in this dataset correspond to null values (as this is the case in the other columns too). So, we are now eliminating these rows.

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
df
```

Now we will format the column. We will get rid of everything but the numbers, and in the end we will create 3 different columns: 'min_salary', 'max_salary' and 'avg_salary. We will do so, because parsing the column in such a way makes the analysis process easier. We can also see that some of the rows contain 'Employer Provided Salary' and some others contain 'Per Hour', so we will store this information before formatting, in the form of a boolean column.

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary' in x.lower() else 0)
```

```python
df['hourly'].value_counts()
```

```python
df['employer_provided'].value_counts()
```

```python
#removing '(Glassdor est.)'
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary
```

```python
#removing 'K' and '$'
minus_Kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
minus_Kd
```

```python
#since we created a boolean column to store the information about the type of salary 'per hour' and 'employer provided salary'
#now we can eliminate these characters
minus_hr_eps = minus_Kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
minus_hr_eps
```

```python
#splitting the column into 'min_salary' and 'max_salary' and creating a calculated column 'avg_salary'
df['min_salary'] = minus_hr_eps.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = minus_hr_eps.apply(lambda x: int(x.split('-')[1]))
df['avg_salary'] = (df['max_salary'] + df['min_salary'])/2
```

```python
df['min_salary'].value_counts().sort_values().head(50)
```

Now we can convert the 'per hour' salary rows into annual salary rows. To do so, we should multiply the hourly wage by 2000, but since our salary numbers are already stored as thousands (there was a 'K' in the dataset), we just need to multiply the hourly wage by 2.

```python
df['min_salary'] = df.apply(lambda x: x.min_salary*2 if x.hourly ==1 else x.min_salary, axis =1)
df['max_salary'] = df.apply(lambda x: x.max_salary*2 if x.hourly ==1 else x.max_salary, axis =1)
```

```python
df[df.hourly ==1][['hourly','min_salary','max_salary']]
```

### Formatting 'Job Title'

```python
df['Job Title'].value_counts().head(30)
```

As we can see the column 'Job Title' includes many different entries for the same title. We will simplify this, and any row that contains a 'data scientist' will be just passed as 'data scientist' and so on for the other titles. Since there are also multiple way to refer to senior or junior positions, we will format that too.

```python
def job_title_simpler(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'                         #machine learning engineer
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'   
```

```python
df['job_simp'] = df['Job Title'].apply(job_title_simpler)
df['job_simp'].value_counts()
```

```python
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
df['seniority'] = df['Job Title'].apply(seniority)
df.seniority.value_counts()
```

### Formatting 'Company Name' 

Here we will just need to get rid of the last 3 characters from the 'Company Name' values, but only for those who have a correct value in the 'Rating' column.

```python
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-3], axis = 1)
df['company_txt']
```

```python
df['company_txt'] = df.company_txt.apply(lambda x: x.replace('\r\n', ''))
df['company_txt'].value_counts().head()
```

### Extract 'State' from 'Location'

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df['job_state'].value_counts()
```

We can see that one value in the column 'job_state' is Los Angeles, we are going to convert that into CA.

```python
df['job_state']= df.job_state.apply(lambda x: x.strip() if x.strip().lower() != 'los angeles' else 'CA')
df.job_state.value_counts()
```

Now we want to add another boolean column to check whether the location of a job corresponds to the headquarters of the company.

```python
df['same_state'] = df.apply(lambda x: 1 if x['Location'] == x['Headquarters'] else 0,axis = 1)
df['same_state'].value_counts()
```

### Extracting 'company_age' from 'Founded'
